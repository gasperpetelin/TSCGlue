```python
from time import perf_counter

import polars as pl
from aeon.classification.convolution_based import (
    RocketClassifier,
)
from aeon.classification.dictionary_based import REDCOMETS, BOSSEnsemble
from aeon.classification.interval_based import (
    DrCIFClassifier,
)
from sklearn.metrics import accuracy_score

from autotsc import utils

```

```python
X_train, y_train, X_test, y_test = utils.load_dataset("FacesUCR")
X_train.shape
```

```python
res_rc = []
```

```python
for i, n_trees in enumerate([1, 1, 20, 50, 100, 200, 500, 800, 1000]):
    model = REDCOMETS(n_jobs=-1, n_trees=n_trees)
    start_time = perf_counter()
    model.fit(X_train, y_train)
    end_time = perf_counter()
    fit_time = end_time - start_time
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    if i == 0:
        continue
    res_rc.append(
        {"param": n_trees, "accuracy": accuracy, "fit_time": fit_time, "model": "REDCOMETS"}
    )
```

```python
for i, n_kernels in enumerate(
    [100, 100, 200, 500, 1000, 2000, 5000, 10000, 20000, 50000, 100000, 200000, 500000]
):
    model = RocketClassifier(n_jobs=-1, n_kernels=n_kernels)
    start_time = perf_counter()
    model.fit(X_train, y_train)
    end_time = perf_counter()
    fit_time = end_time - start_time
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    if i == 0:
        continue
    res_rc.append(
        {"param": n_kernels, "accuracy": accuracy, "fit_time": fit_time, "model": "MiniRocket"}
    )
```

```python
for i, n_estimators in enumerate([1, 3, 5, 10, 20, 50, 100, 200]):
    model = DrCIFClassifier(n_jobs=-1, random_state=42, n_estimators=n_estimators)
    start_time = perf_counter()
    model.fit(X_train, y_train)
    end_time = perf_counter()
    fit_time = end_time - start_time
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    if i == 0:
        continue
    res_rc.append(
        {"param": n_estimators, "accuracy": accuracy, "fit_time": fit_time, "model": "DrCIF"}
    )
```

```python
df = pl.DataFrame(res_rc)
df
```

```python
import seaborn as sns

sns.lineplot(data=df, x="fit_time", y="accuracy", hue="model", style="model", markers=True)

# df_rc = df.filter(pl.col('model') == 'REDCOMETS').sort('fit_time')
# plt.plot(df_rc['fit_time'], df_rc['accuracy'], marker='.')
# df_roc = df.filter(pl.col('model') == 'MiniRocket').sort('fit_time')
# plt.plot(df_roc['fit_time'], df_roc['accuracy'], marker='.')
# df_drc = df.filter(pl.col('model') == 'DrCIF').sort('fit_time')
# plt.plot(df_drc['fit_time'], df_drc['accuracy'], marker='.')
```
