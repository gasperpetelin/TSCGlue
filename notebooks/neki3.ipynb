```python
from aeon.benchmarking import resampling
from aeon.classification.convolution_based import MultiRocketHydraClassifier
from sklearn.metrics import accuracy_score

from autotsc import utils
from run_stacking import *
```

```python
X_train, y_train, X_test, y_test = utils.load_dataset("DistalPhalanxTW")
X_train, y_train, X_test, y_test = resampling.stratified_resample_data(
    X_train, y_train, X_test, y_test
)
print(f"Train size: {X_train.shape}, Test size: {X_test.shape}")
```

```python
model = MultiRocketHydraClassifier(n_jobs=-1)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
# split train into train and val
for _ in range(10):
    from sklearn.model_selection import train_test_split

    X_train_split, X_val, y_train_split, y_val = train_test_split(
        X_train, y_train, test_size=0.2, stratify=y_train
    )
    model = MultiRocketHydraClassifier(n_jobs=-1)
    model.fit(X_train_split, y_train_split)
    y_val_pred = model.predict(X_val)
    val_accuracy = accuracy_score(y_val, y_val_pred)
    print(f"Validation Accuracy: {val_accuracy}")
```
