```python
from aeon.pipeline import make_pipeline as aeon_make_pipeline
from autotsc import transformers, utils
from aeon.classification.convolution_based import MultiRocketHydraClassifier
from autotsc.gpu_models import LightningSGDClassifier, MRHydraClassifier
from sklearn.metrics import accuracy_score
from aeon.transformations.collection.convolution_based import MultiRocket
from aeon.transformations.collection.convolution_based._hydra import HydraTransformer
from autotsc.gpu_models import LightningSGDClassifier, MRHydraClassifier
import numpy as np
from sklearn.linear_model import RidgeClassifierCV
from sklearn.feature_selection import SelectKBest, f_classif, mutual_info_classif
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import VarianceThreshold

dataset = "GestureMidAirD2"
X_train, y_train, X_test, y_test = utils.load_dataset(dataset)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
e = RidgeClassifierCV(alphas=np.logspace(-3, 3, 10))
m = MRHydraClassifier(estimator=e, random_state=42, n_jobs=-1)
m.fit(X_train, y_train)
pred = m.predict(X_test)
print("Accuracy:", accuracy_score(y_test, pred))
```

```python
for k in [1000, 2000, 3000, 4000, 5000, 10_000, 50_000]:
    e1 = Pipeline(
        [
            ("var", VarianceThreshold()),
            ("select", SelectKBest(f_classif, k=k)),
            ("clf", RidgeClassifierCV(alphas=np.logspace(-3, 3, 10))),
        ]
    )
    m = MRHydraClassifier(estimator=e1, random_state=42, n_jobs=-1)
    m.fit(X_train, y_train)
    pred = m.predict(X_test)
    print("Accuracy:", accuracy_score(y_test, pred), "k=", k)
```

```python
from sklearn.linear_model import SGDClassifier


e = SGDClassifier(
    loss="hinge",
    alpha=1e-4,
    max_iter=200,
    tol=1e-4,
    learning_rate="optimal",
    early_stopping=False,
    #validation_fraction=0.1,
    n_iter_no_change=10,
    average=True,
    random_state=42,
    verbose=0
)
m = MRHydraClassifier(estimator=e, random_state=42, n_jobs=-1)
m.fit(X_train, y_train)
pred = m.predict(X_test)
print("Accuracy:", accuracy_score(y_test, pred))
```
