```python
import numpy as np
from aeon.transformations.collection.interval_based import RandomIntervals
from tscglue import data_loader
from tscglue.interval_models import _FastRandomIntervals

X_train, y_train, X_test, y_test = data_loader.load_fold("Fish", fold=0)
print(f"train={X_train.shape}  test={X_test.shape}")
```

```python
SEED = 42
N_INTERVALS = 650

aeon_ri = RandomIntervals(n_intervals=N_INTERVALS, min_interval_length=3, random_state=SEED, n_jobs=1)
fast_ri = _FastRandomIntervals(n_intervals=N_INTERVALS, min_interval_length=3, random_state=SEED)

fast_ri.fit(X_train)
aeon_ri.fit(X_train)
aeon_train = aeon_ri.transform(X_train)
fast_train = fast_ri.transform(X_train)

aeon_test = aeon_ri.transform(X_test)
fast_test = fast_ri.transform(X_test)

print(f"aeon train shape : {aeon_train.shape}")
print(f"fast train shape : {fast_train.shape}")
print(f"shapes match     : {aeon_train.shape == fast_train.shape}")
print()
print(f"train allclose   : {np.allclose(aeon_train, fast_train)}")
print(f"train max diff   : {np.max(np.abs(aeon_train - fast_train)):.2e}")
print()
print(f"test allclose    : {np.allclose(aeon_test, fast_test)}")
print(f"test max diff    : {np.max(np.abs(aeon_test - fast_test)):.2e}")
```

```python
# Also check intervals are identical
aeon_intervals = [(s, e, d, dil) for s, e, _, d, dil in aeon_ri.intervals_]
fast_intervals = [(s, e, d, dil) for s, e, _, d, dil in fast_ri._ri.intervals_]

print(f"n_intervals aeon : {len(aeon_intervals)}")
print(f"n_intervals fast : {len(fast_intervals)}")
print(f"intervals match  : {aeon_intervals == fast_intervals}")
```
