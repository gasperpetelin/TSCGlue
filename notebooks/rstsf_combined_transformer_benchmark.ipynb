```python
from tscglue.interval_models import RSTSFCombinedTransformer
from tscglue import data_loader

X_train, y_train, X_test, y_test = data_loader.load_fold('Fish', fold=5)
print(f'X_train: {X_train.shape}')
```

```python
t = RSTSFCombinedTransformer(verbose=True, n_intervals_random=600, n_intervals_unsupervised=50, random_state=42, n_jobs=-1)
Xt_train = t.fit_transform(X_train)
Xt_test  = t.transform(X_test)
print(f'Xt_train: {Xt_train.shape} ({Xt_train.nbytes / 1024**2:.1f} MB)')
print(f'Xt_test:  {Xt_test.shape}  ({Xt_test.nbytes / 1024**2:.1f} MB)')
```
