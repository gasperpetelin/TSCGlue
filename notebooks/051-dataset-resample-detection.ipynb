```python
import numpy as np
from aeon.benchmarking import resampling
from aeon.classification.convolution_based import MultiRocketHydraClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

from autotsc import utils
```

```python
dataset = "Worms"
```

```python
X_train1, y_train1, X_test1, y_test1 = utils.load_dataset(dataset)
# X_train2, _, X_test2, _ = resampling.stratified_resample_data(X_train1, y_train1, X_test1, y_test1)
X_train1, y_train1, X_test1, y_test1 = resampling.stratified_resample_data(
    X_train1, y_train1, X_test1, y_test1
)
```

```python
# join all into 1 dataset to check if i can predict resamples data
X_all = np.concatenate([X_train1, X_test1], axis=0)
y_all = np.array([0] * len(X_train1) + [1] * len(X_test1))
print(X_train1.shape, X_test1.shape, X_all.shape, y_all.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_all, y_all, test_size=0.2, stratify=y_all)
```

```python
model = MultiRocketHydraClassifier(n_jobs=-1)
# model = Catch22Classifier()
# model = DummyClassifier(strategy='stratified')
model.fit(X_train, y_train)
y_val_pred = model.predict(X_test)
val_accuracy = accuracy_score(y_test, y_val_pred)
val_accuracy

from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_val_pred)
f1
```
