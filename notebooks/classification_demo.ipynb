# TSCGlue Classification Demo

Trains `TSCGlueClassifier` on a small UCR dataset (GunPoint: 50 train / 150 test, binary, length 150).

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score

from tscglue import utils
from tscglue.models import TSCGlueClassifier
```

```python
DATASET = "GunPoint"
```

```python
X_train, y_train, X_test, y_test = utils.load_dataset(DATASET)
print(f"Train: {X_train.shape}, Test: {X_test.shape}")
print(f"Classes: {np.unique(y_train)}")
```

```python
clf = TSCGlueClassifier(n_jobs=4, verbose=1)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
print(f"Test accuracy: {accuracy_score(y_test, y_pred):.4f}")
```
