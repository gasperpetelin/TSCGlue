```python
from aeon.transformations.collection.interval_based import QUANTTransformer
from aeon.transformations.collection.convolution_based import MultiRocket
from aeon.transformations.collection.convolution_based._hydra import HydraTransformer
from tscglue.data_loader import load_fold, DATA_DIR
import os
```

```python
datasets = sorted(d for d in os.listdir(DATA_DIR) if os.path.isdir(os.path.join(DATA_DIR, d)))
print(f"{len(datasets)} datasets")
datasets[:10]
```

```python
ds = datasets[0]
X_train, y_train, X_test, y_test = load_fold(ds, fold=0)
print(f"Dataset: {ds}")
print(f"X_train shape: {X_train.shape}")
```

```python
qt = QUANTTransformer()
mr = MultiRocket(random_state=0)
hy = HydraTransformer(random_state=0)

X_quant = qt.fit_transform(X_train)
X_rocket = mr.fit_transform(X_train)
X_hydra = hy.fit_transform(X_train)

print(f"Quant features:      {X_quant.shape[1]:>6}")
print(f"MultiRocket features:{X_rocket.shape[1]:>6}")
print(f"Hydra features:      {X_hydra.shape[1]:>6}")
```

```python
# Check a few datasets to see how feature count varies
for ds in datasets[:10]:
    X_train, y_train, _, _ = load_fold(ds, fold=0)
    qt = QUANTTransformer()
    mr = MultiRocket(random_state=0)
    hy = HydraTransformer(random_state=0)
    n_quant = qt.fit_transform(X_train).shape[1]
    n_rocket = mr.fit_transform(X_train).shape[1]
    n_hydra = hy.fit_transform(X_train).shape[1]
    print(f"{ds:40s}  X={str(X_train.shape):20s}  quant={n_quant:>6}  rocket={n_rocket:>6}  hydra={n_hydra:>6}")
```
