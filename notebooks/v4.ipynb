```python
s = StackerV4(random_state=270, n_repetitions=1, k_folds=10, time_limit_in_seconds=None)
s.fit(X_train, y_train)
```

```python
import pickle
import time

import polars as pl

# Assuming Xt is a numpy array or dataframe

# Pickle serialize
start = time.time()
pickle.dumps(Xt)
print(f"Pickle serialize: {time.time() - start:.3f}s")

# Parquet write
start = time.time()
pl.DataFrame(Xt).write_parquet("/tmp/test.parquet", compression="uncompressed")
print(f"Parquet write: {time.time() - start:.3f}s")

# Numpy save
import numpy as np

start = time.time()
np.save("/tmp/test.npy", Xt)
print(f"Numpy save: {time.time() - start:.3f}s")
```
