```python
from aeon.pipeline import make_pipeline as aeon_make_pipeline
from autotsc import transformers, utils
from aeon.classification.convolution_based import MultiRocketHydraClassifier
from autotsc.gpu_models import LightningSGDClassifier, MRHydraClassifier
from sklearn.metrics import accuracy_score

dataset = "GestureMidAirD2"
X_train, y_train, X_test, y_test = utils.load_dataset(dataset)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
m = MRHydraClassifier(n_jobs=-1)
m.fit(X_train, y_train)
print(m.hydra_time_)
print(m.multirocket_time_)
print(m.classifier_time_)
```

```python
pred = m.predict(X_test)
print("Accuracy:", accuracy_score(y_test, pred))
```

```python
estimator = LightningSGDClassifier(max_epochs=1, verbose=10, batch_size=32)
m = MRHydraClassifier(n_jobs=-1, estimator=estimator)
m.fit(X_train, y_train)
print(m.hydra_time_)
print(m.multirocket_time_)
print(m.classifier_time_)
```

```python
pred = m.predict(X_test)
print("Accuracy:", accuracy_score(y_test, pred))
```
