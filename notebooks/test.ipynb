```python
from sklearn.metrics import accuracy_score

from autotsc import utils
from autotsc.models import *
```

```python
X_train, y_train, X_test, y_test = utils.load_dataset("Worms")
```

```python
with utils.ray_init_or_reuse(num_cpus=24, resources={"meta": 100}, ignore_reinit_error=True):
    model = AutoTSCModel(verbose=1, n_jobs=-1, n_gpus=-1, model_selection="fast")
    model.fit(X_train, y_train)
```

```python
model.oof_predictions_
```

```python
mm = Ensemble(RidgeClassifierCVWithProba(alphas=np.logspace(-3, 3, 10)))
mm.fit_predict_proba_cv(model.oof_predictions_.to_numpy(), y_train, model.folds_)
```

```python
m = RayCrossValidationWrapper(Catch22Classifier(n_jobs=-1))
folds = utils.get_folds(X_train, y_train, n_splits=10)
probs = m.fit_predict_proba(X_train, y_train, cv_splits=folds)
probs
```

```python
pred = m.predict(X_test)
accuracy_score(y_test, pred)
```
