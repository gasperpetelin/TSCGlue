```python
import polars as pl


# ----------------------------------------------------------------------
# STEP 1 — Load model preds and labels from your Polars DF
# ----------------------------------------------------------------------
def load_models_from_df(df: pl.DataFrame):
    """
    Convert Polars rows into Python objects suitable for ensemble selection.
    """
    models = []
    for row in df.iter_rows(named=True):
        models.append(
            {
                "model_id": row["model_id"],
                "probs": np.array(row["fold_probabilities"], dtype=float),
                "true": np.array(row["true_labels"]),
                "val_acc": row["validation_accuracy"],
            }
        )
    return models


# ----------------------------------------------------------------------
# STEP 2 — Accuracy helper — CORRECT
# ----------------------------------------------------------------------
def accuracy_from_probs(probs: np.ndarray, true_labels: np.ndarray) -> float:
    """
    Convert probabilities to class predictions and compute accuracy.
    Classes are ordinal (sorted unique values).
    """
    # Get sorted unique classes (ordinal order)
    unique_classes = sorted(np.unique(true_labels).astype(int))

    # argmax gives index in probs columns, which maps to sorted unique classes
    pred_indices = probs.argmax(axis=1)  # [0, 1, 0, ...]
    preds = np.array([str(unique_classes[idx]) for idx in pred_indices])

    return (preds == true_labels).mean()


# ----------------------------------------------------------------------
# STEP 3 — Caruana Ensemble Selection
# ----------------------------------------------------------------------
def caruana_ensemble_selection(models, iterations=50):
    """
    Implements the original Caruana et al. (2004/2006) ensemble selection.
    Ensemble uses weighted average of model probabilities.
    """

    # Stack prediction matrices
    preds = np.stack([m["probs"] for m in models])
    true = models[0]["true"]
    n_models = len(models)

    # Step 1: Pick best single model
    accuracies = np.array([accuracy_from_probs(preds[i], true) for i in range(n_models)])

    selected = [accuracies.argmax()]
    ensemble_sum = preds[selected].sum(axis=0)
    best_acc = accuracies[selected[0]]

    print(f"Initial model: {models[selected[0]]['model_id']} acc={best_acc:.4f}")

    # Step 2: Greedy selection with replacement
    for t in range(iterations):
        best_candidate = None
        best_candidate_acc = best_acc

        for i in range(n_models):
            new_sum = ensemble_sum + preds[i]
            new_mean = new_sum / (len(selected) + 1)

            acc = accuracy_from_probs(new_mean, true)

            if acc > best_candidate_acc:
                best_candidate_acc = acc
                best_candidate = i

        if best_candidate is None:
            print(f"Iter {t + 1}: No improvement possible. Stopping.")
            break

        selected.append(best_candidate)
        ensemble_sum += preds[best_candidate]
        best_acc = best_candidate_acc

        print(f"Iter {t + 1}: Added {models[best_candidate]['model_id']} => acc={best_acc:.4f}")

    final_probs = ensemble_sum / len(selected)

    return selected, final_probs, best_acc
```

```python
# USAGE
# ======================================================================
# Filter to stacking_level == 0 models
summary_base = summary.filter(pl.col("stacking_level") == 0)

models = load_models_from_df(summary_base)

selected, final_probs, final_acc = caruana_ensemble_selection(models, iterations=50)

print("\nSelected model IDs (with repetition):")
print([models[i]["model_id"] for i in selected])

print(f"\nFinal ensemble accuracy: {final_acc:.4f}")
print(f"Ensemble size: {len(selected)} models")
```
